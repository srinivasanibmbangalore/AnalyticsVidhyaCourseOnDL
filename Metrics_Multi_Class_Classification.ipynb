```
from sklearn import metrics
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
```

```
C="Cat"
D="Dog"
F="Fox"
y_truth=[C,C,C,C,C,C,F,F,F,F,F,F,F,F,F,F,D,D,D,D,D,D,D,D,D]
y_pred= [C,C,C,C,D,F,C,C,C,C,C,C,D,D,F,F,C,C,C,D,D,D,D,D,D]
```

```
metrics.confusion_matrix(y_truth,y_pred)
```

#### Note the diagonals of the Confusion Matrix : They are the true positive for each of the label - i.e 4 for Cats, 6 for Dogs and 2 for Fox



```
print(metrics.classification_report(y_truth,y_pred,digits=4))
```

Total Number of Test Examples of any class would be sum of corresponding row. i.e TP+FN for that class

Total Number of FN for a class is sum of values in the corresponding row (excluding TP)

Precision=TP/(TP+FP)

Precision(Cat) = Number of True Prediction of Cat/(Number of True Prediction of Cat + Number of False Prediction of Cats)

Precision(Dog)=Number of True Prediction of Dogs/(Number of True Prediction of Dogs + Number of False Prediction of Dogs)

Precision(Fox)=Number of True Prediction of Fox/(Number of True Prediction of Fox + Number of False Prediction of Fox)

```
p_cat=4/13
print(p_cat)
```

```
p_fox=2/3
print(p_fox)
```

Recall = (TP/TP+FN)
Recall(Cat)=Number of True Prediction of Cat/(Number of True Prediction of Cat + Number of Actual Cats but not predicted as Cat)

Recall(Dog)=Number of True Prediction of Dog/(Number of True Prediction of Dog + Number of Actual Dogs but not predicted as Dogs)


```
r_cat=4/(4+2)
print(r_cat)
```

```
r_dog=(6/(6+3))
print(r_dog)
```

```
r_fox=2/(2+8)
print(r_fox)
```
